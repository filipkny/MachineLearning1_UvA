Write a method `gen_sine(N)` that generates toy data like in fig 1.2 of Bishop's book. The method should have a parameter $N$, and should return $N$-dimensional vectors $\bx$ and $\bt$, where $\bx$ contains evenly spaced values from 0 to (including) 2$\pi$, and the elements $t_i$ of $\bt$ are distributed according to:

$$t_i \sim \mathcal{N}(\mu_i, \sigma^2)$$

where $x_i$ is the $i$-th elements of $\bf{x}$, the mean $\mu_i = \sin(x_i)$ and the standard deviation $\sigma = 0.25$.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
begin_sine = 0
termination_sine = 2.0*np.pi

# interpeting this as sigma^2 as if we squares 0.25 the variance will be too small to be interesting
var = 0.25

def gen_sine_general(n, variance, start, end):
    """ a more general version of gen_sine that is used to generate data with specified interval and variance"""
    
    # generated x linspace
    x = np.linspace(start, end, n)
    
    # find corresponding t
    t = []
    for x_ in x:
        mean = np.sin(x_)
        t.append(np.random.normal(mean, variance))
    return np.array(x), np.array(t)

def gen_sine(n):
    """ generates dummy data from a sine function """
    
    return gen_sine_general(n, var, begin_sine, termination_sine)
```

### 1.2 Polynomial regression (10 points)

Write a method `fit_polynomial(x, t, M)` that finds the maximum-likelihood solution of an _unregularized_ $M$-th order polynomial for some dataset `x`. The error function to minimize w.r.t. $\bw$ is:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt)$

where $\bPhi$ is the _feature matrix_ (or _design matrix_) as explained in Bishop's book at section 3.1.1, $\bt$ is the vector of target values. Your method should return a vector $\bw$ with the maximum-likelihood parameter estimates, as well as the _feature matrix_ $\bPhi$.

```python
def phi(input_x, power):
    """ returns an input to a certain power"""
    
    return input_x**power

def designmatrix(x, M):
    """ returns the feature matrix """
    
    return np.matrix([[  phi(x_, m) for m in range(M+1) ]  for x_ in x])

def fit_polynomial(x, t, M):
    """ fits a polynomial with least squares in closed form, returns weights and featurematrix"""

    Phi = designmatrix(x, M)    
    pseudo_inv = np.linalg.pinv(Phi)
    w_ml = np.dot(pseudo_inv, t)
    
    # reshape is needed to get right object for the test
    return np.array(w_ml.T).reshape((M+1,)), Phi
```

### 1.4 Regularized linear regression (15 points)

a) (10 points) Write a method `fit_polynomial_reg(x, t, M, lamb)` that fits a _regularized_ $M$-th order polynomial to the periodic data, as discussed in the lectures, where `lamb` is the regularization term _lambda_. (Note that 'lambda' cannot be used as a variable name in Python since it has a special meaning). The error function to minimize w.r.t. $\bw$:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt) + \frac{\lambda}{2} \mathbf{w}^T \mathbf{w}$

For background, see section 3.1.4 of Bishop's book.

The function should return $\bw$ and $\bPhi$.

```python
def fit_polynomial_reg(x, t, M, lamb):
    """ fit regularized least squares"""
    
    # get featurematrix
    D_M = designmatrix(x, M)    

    # get term to be inverted
    step_a = np.dot(D_M.T, D_M) + (np.dot( lamb, np.eye(M+1)))

    # inverse previous term and do dot product with feature matrix
    step_b = np.dot(np.linalg.inv(step_a), D_M.T,)
    
    # find weights by taking dot product with targets
    w_ml = np.dot(step_b, t)
    
    # return weights and featurematrix
    return np.array(w_ml.T).reshape((M+1,)), D_M
```

b) (10 points) Now write a method find_best_m_and_lamb(x, t) that finds the best values for $M$ and $\lambda$. The method should return the best $M$ and $\lambda$. To get you started, here is a method you can use to generate indices of cross-validation folds.

```python
def kfold_indices(N, k):
    all_indices = np.arange(N,dtype=int)
    np.random.shuffle(all_indices)
    idx = [int(i) for i in np.floor(np.linspace(0,N,k+1))]
    train_folds = []
    valid_folds = []
    for fold in range(k):
        valid_indices = all_indices[idx[fold]:idx[fold+1]]
        valid_folds.append(valid_indices)
        train_folds.append(np.setdiff1d(all_indices, valid_indices))
    return train_folds, valid_folds
```

```python
def find_best_m_and_lamb(x, t):
    """ grid search over 10x10 values of the order of polynomial and regularizer values """
    
    # predefined
    k=5
    
    # fold 
    train_folds, valid_folds = kfold_indices(len(x), k)
    
    # init for keeping track of best
    prediction_errors = {key : {key2: sys.maxsize for key2 in range(-10,1)} for key in range(0,11)}
    best = sys.maxsize
    argbest=(0,0)
    
    # do grid search
    for m in prediction_errors:
        for lamb in prediction_errors[m]:
            error = 0
            for kth in range(len(train_folds)):
                exp = np.exp(lamb)
                error += pred_error(x[train_folds[kth]],
                                    x[valid_folds[kth]],
                                    t[train_folds[kth]],
                                    t[valid_folds[kth]],
                                    m,
                                    exp)
            error /= float(len(train_folds))
            if (error < best):
                best = error
                argbest = (m,exp)
            prediction_errors[m][lamb] = error
            
    return argbest

```

### 1.6 Why grid search? (5 points)

Grid search is an commonly-used technique to tune hyper-parameters in a model.
Considering the case described in the previous step of this assignment, running a grid search over the possible parameter values (10 possible values for both $M$ and $\lambda$), results in two nested loops exploring $10 \times 10 = 100$ different configurations for the model. 

a) (3 points) Why do we want to optimize by changing the two hyperparameters at the same time, and not in a sequential way? We could initialise all parameters randomly, fix one parameter at a time and iterate over the other, resulting in only $10 + 10 = 20$ experiments!

**The influence of both parameters in the accuracy of the model is not independant. The fact that lambda = e^-5 works better than e^-6 for m = 3 doesn't mean that you can expect the same for m = 4. Therefore, there needs to be some structure in the search for the optimal parameters. By doing two nested loops where we change the hyper parameters using constant increments allows us to estimate much better how the error landscape looks like and thus ends up being a more efficient search method.**

For more complex models, the number of combinations easily explodes with the number of parameters. For example, with 5 parameters we would run $10 \times 10 \times 10 \times 10 \times 10 = 100,000$ experiments.

b) (2 points) Try to think or find in literature one alternative to grid search to tune hyper-parameters more efficiently. Explain very briefly (2-3 lines max) how this method avoids the combinatorial explosion we have see in grid search.

**A possibility is to use an Evolutionary Algorithm with an adapative stepsize when changing the parameters during the search. In the beginning of said search, it is natural to have higher step sizes since the error landscape is completely unknown. As the search advances, this step size could be make smaller to exploit clear peaks of the landscape**

### 2.2 Compute Posterior (15 points)

You're going to implement a Bayesian linear regression model, and fit it to the periodic data. Your regression model has a zero-mean isotropic Gaussian prior over the parameters, governed by a single (scalar) precision parameter $\alpha$, i.e.:

$$p(\bw \;|\; \alpha) = \mathcal{N}(\bw \;|\; 0, \alpha^{-1} \bI)$$

The covariance and mean of the posterior are given by:

$$\bS_N= \left( \alpha \bI + \beta \bPhi^T \bPhi \right)^{-1} $$
$$\bm_N = \beta\; \bS_N \bPhi^T \bt$$

where $\alpha$ is the precision of the predictive distribution, and $\beta$ is the noise precision. 
See MLPR chapter 3.3 for background.

Write a method `fit_polynomial_bayes(x, t, M, alpha, beta)` that returns the mean $\bm_N$ and covariance $\bS_N$ of the posterior for a $M$-th order polynomial. In addition it should return the design matrix $\bPhi$. The arguments `x`, `t` and `M` have the same meaning as in question 1.2.

```python
def fit_polynomial_bayes(x, t, M, alpha, beta):
    """ finds the distribution over weights in bayesian setting """
    
    # feature matrix
    Phi = designmatrix(x, M)
    Phi_squared = np.dot(Phi.T, Phi)
    
    # find covariance of weights
    S = np.linalg.inv(np.dot(alpha, np.eye(len(Phi_squared)))+ np.dot(beta, Phi_squared))
    
    # find means of weights
    m = np.dot(beta, np.linalg.multi_dot([S, Phi.T, t]))
    
    return np.array(m).reshape((M+1,)), S, Phi

# for testing
N = 10
x = np.linspace(-1, 1, N)
t = 0.3*np.square(x) + 2.5
M = 2
alpha = 0.6
beta = 16
m, S, Phi = fit_polynomial_bayes(x, t, M, alpha, beta)
```

### 2.3 Prediction (10 points)

The predictive distribution of Bayesian linear regression is:

$$ p(t \;|\; \bx, \bt, \alpha, \beta) = \mathcal{N}(t \;|\; \bm_N^T \phi(\bx), \sigma_N^2(\bx))$$

$$ \sigma_N^2 = \frac{1}{\beta} + \phi(\bx)^T \bS_N \phi(\bx) $$

where $\phi(\bx)$ are the computed features for a new datapoint $\bx$, and $t$ is the predicted variable for datapoint $\bx$. 

Write a function that `predict_polynomial_bayes(x, m, S, beta)` that returns the predictive mean, variance and design matrix $\bPhi$ given a new datapoint `x`, posterior mean `m`, posterior variance `S` and a choice of model variance `beta`.

```python
def expand_phi(x, m):
    """  expands an x with basis functions phi to order m"""
    
    return np.array([phi(x, m_) for m_ in range(0,m)]).T

def predict_polynomial_bayes(x, m, S, beta):
    """ gives the distribution of predictions given all new data x, and a mean and covariance over the weights"""
    
    # find order
    M = len(m)
    
    # get feature matrix
    Phi = designmatrix(x, M-1)

    # get a predictive variance per datapoint
    sigma = np.array([(1.0/beta) + np.linalg.multi_dot([expand_phi(x_, M).T, S, expand_phi(x_, M)]) for x_ in x])
    
    # get a predictive mean per datapoint
    mean = np.array([np.dot(m.T, expand_phi(x_, M)) for x_ in x])
    
    return mean, sigma, Phi
```
